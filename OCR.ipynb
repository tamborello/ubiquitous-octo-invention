<h1>OCR the Images Extracted from the JFK Files PDFs</h1>

```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract

import os
from utils.whoopsie import whoopsie
```

```python
def ocr_doc (basedir, dir) :
    doc = ''
    try:
        dirlist = os.listdir(basedir + dir)
    except Exception as exc:
        whoopsie(str(exc))
    for img in dirlist:
        doc = doc + pytesseract.image_to_string(Image.open(basedir + dir + '/' + img))

    with open('docs/2018_txt_dump/' + dir + '.txt', 'w') as f:
        f.write(doc)
```

```python
ocr_doc('docs/2018_img_dump/104-10086-10009/')
```

```python
# def process_corpus ():
basedir = 'docs/2018_img_dump/'
for dir in os.listdir(basedir):
    try:
        print(dir)
        ocr_doc(basedir, dir)
    except Exception as exc:
        whoopsie(str(exc))
```
